### Model Training

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(df_1[['f1']], df_1['y'])
```

```python
print(reg.coef_, reg.intercept_)
```

```python
pred = reg.predict(df_1[['f1']])
```

```python
plt.scatter(df_1['f1'], df_1['y'])

plt.plot(df_1['f1'], pred, 'r')

plt.show()
```

### Feature Engineering - Learning Polynomial Features

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(df_1[['f1']])
X_poly
```

```python
transformed_df_1 = pd.DataFrame(X_poly, columns=['f1_0', 'f1_1', 'f1_2'])

transformed_df_1['y'] = df_1['y']

transformed_df_1.head()
```

```python
sns.pairplot(transformed_df_1)
```

```python
reg_poly = LinearRegression()
reg_poly.fit(transformed_df_1[['f1_0', 'f1_1', 'f1_2']], transformed_df_1['y'])
```

```python
print(reg_poly.coef_, reg_poly.intercept_)
```

```python
pred_poly = reg_poly.predict(transformed_df_1[['f1_0', 'f1_1', 'f1_2']])
```

```python
plt.scatter(df_1['f1'], df_1['y'])

plt.plot(df_1['f1'], pred_poly, 'r')

plt.show()
```

### Other Regression Models - Lasso, Ridge, ElasticNet

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import ElasticNet
```

```python
reg_lasso = Lasso(alpha=1.0)
reg_lasso.fit(transformed_df_1[['f1_0', 'f1_1', 'f1_2']], transformed_df_1['y'])
```

```python
pred_lasso = reg_lasso.predict(transformed_df_1[['f1_0', 'f1_1', 'f1_2']])
```

```python
plt.scatter(df_1['f1'], df_1['y'])

plt.plot(df_1['f1'], pred_lasso, 'r')

plt.show()
```

```python
reg_ridge = Ridge(alpha=1.0)
reg_ridge.fit(transformed_df_1[['f1_0', 'f1_1', 'f1_2']], transformed_df_1['y'])
```

```python
pred_ridge = reg_ridge.predict(transformed_df_1[['f1_0', 'f1_1', 'f1_2']])
```

```python
plt.scatter(df_1['f1'], df_1['y'])

plt.plot(df_1['f1'], pred_ridge, 'r')

plt.show()
```

```python
reg_elastic = ElasticNet(alpha=1.0)
reg_elastic.fit(transformed_df_1[['f1_0', 'f1_1', 'f1_2']], transformed_df_1['y'])
```

```python
pred_elastic = reg_elastic.predict(transformed_df_1[['f1_0', 'f1_1', 'f1_2']])
```

```python
plt.scatter(df_1['f1'], df_1['y'])

plt.plot(df_1['f1'], pred_elastic, 'r')

plt.show()
```
